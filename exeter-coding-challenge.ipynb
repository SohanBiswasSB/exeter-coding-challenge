```python
import time
import csv
import os
import psutil
```

```python
start_time = time.time()
```

```python
french_dictionary={}
french_file=open('french_dictionary.csv','r')
csv_reader=csv.reader(french_file)
for row in csv_reader:
    french_dictionary[row[0]]=row[1]
```

```python
find_word_file=open('find_words.txt', 'r')
all=find_word_file.read()
find_words_list=all.split()
find_word_file.close()
```

```python
count=dict()
with open('t8.shakespeare.txt','r') as file:
    filedata=file.read()
shakespeare_file=open('t8.shakespeare.txt', 'r')
```

```python

for line in shakespeare_file:
    words=line.split()
    for word in words:
        if word.lower() in find_words_list:
            if word.lower() in count:
                count[word.lower()]+=1
            elif word.lower().strip(",") in count:
                count[word.lower().strip(",")]+=1
            elif word.lower().strip("'s") in count:
                count[word.lower().strip("'s")]+=1
            elif word.lower().strip("-") in count:
                count[word.lower().strip("-")]+=1
            elif word.lower().strip("@") in count:
                count[word.lower().strip("@")]+=1
            elif word.lower().strip(";") in count:
                count[word.lower().strip(";")]+=1
            elif word.lower().strip(":") in count:
                count[word.lower().strip(":")]+=1
            elif word.lower().strip(".") in count:
                count[word.lower().strip(".")]+=1
            elif word.lower().strip("?") in count:
                count[word.lower().strip("?")]+=1
            elif word.lower().strip(".'") in count:
                count[word.lower().strip(".'")]+=1
            else:
                count[word.lower()]=1
                filedata=filedata.replace(word,french_dictionary[word.lower()])
                new_word=word+","
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+"'s"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+":"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+"."
                filedata = filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+"-"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+"@"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+";"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+"?"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
                new_word=word+".'"
                filedata=filedata.replace(new_word,french_dictionary[word.lower()])
```

```python
with open('t8.shakespeare.translated.txt','w') as file:
    file.write(filedata)
```

```python
with open('frequency.csv','w',newline='') as csvfile:
    fieldnames=['English Word', 'French Word', 'Frequency']
    thewriter=csv.DictWriter(csvfile, fieldnames=fieldnames)
    thewriter.writeheader()
    for i in count:
        thewriter.writerow({'English Word':i,'French Word':french_dictionary[i],'Frequency':count[i]})
```

```python
stop_time=time.time()-start_time
minutes=stop_time//60
seconds=stop_time%60
print("Time taken to process: %d minutes %d seconds"%(minutes,seconds))
total_memory=psutil.Process(os.getpid()).memory_info().rss/1024 ** 2
print(f"Memory Used: {round(total_memory,2)} MB")
```
